```python
import os
import cv2
import albumentations as A
```

```python
aug_number = 3
```

```python
augmentations = A.Compose([
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.2),
    A.RandomBrightnessContrast(p=0.3),
    A.Rotate(limit=25, p=0.7),
    A.GaussianBlur(blur_limit=(3, 5), p=0.3),
    A.GaussNoise(var_limit=(10.0, 50.0), p=0.3),
])

def augment_image(image):
    return [augmentations(image=image)["image"] for _ in range(aug_number)]

def process_images(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for filename in os.listdir(input_folder):
        if filename.lower().endswith(('.png', '.jpg')):
            img_path = os.path.join(input_folder, filename)
            image = cv2.imread(img_path)
            if image is None:
                continue

            augmented_images = augment_image(image)
            name, ext = os.path.splitext(filename)

            for i, aug_img in enumerate(augmented_images):
                save_path = os.path.join(output_folder, f"{name}_aug{i}{ext}")
                cv2.imwrite(save_path, aug_img)

            print(f"{filename} -> {aug_number} augmentierte Versionen gespeichert")
```

```python
input_folder = "images/Ancistrus_dolichopterus"
output_folder = "aug_images/Ancistrus_dolichopterus"
process_images(input_folder, output_folder)
```
